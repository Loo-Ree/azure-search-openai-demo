```python
import xml.etree.ElementTree as ET
import json
import os

# setup the namespaces
namespaces = {'ns': 'http://docs.oasis-open.org/legaldocml/ns/akn/3.0'}

# specify the directory where the XML files are located
directory = "c:\\Users\\lrepichini\\OneDrive - Microsoft\\Workspace\\SMD-aoai\\data\\type2"

# loop processing each XML file in the directory
for filename in os.listdir(directory):
    # only process XML files
    if filename.endswith(".xml"):
        # parse the XML file
        xml_file = os.path.join(directory, filename)
        tree = ET.parse(xml_file)
        root = tree.getroot()

        # create an empty list to store the json data
        data = []
        
        for chapter in root.findall('.//ns:chapter', namespaces):
            # Extract the heading content
            heading = chapter.find('ns:heading', namespaces)
            heading_content = ''.join(heading.itertext()).strip() if heading is not None else None

            # loop through each doc element in the XML file and select only the tags named <doc/>
            for article in chapter.findall('.//ns:article', namespaces):
                # the name contains the article number
                original_name = article.get('eId')
        
                # Split the original name into a prefix and a number
                prefix, number = original_name.split('_')
        
                # Format the new name
                name = "Art. " + number
                
                # Iterate over all elements in the tree
                for parent in article.iter():
                    
                    # Remove all 'authorialNote' elements that are children of the current element
                    for authorialNote in list(parent):
                        if authorialNote.tag == '{http://docs.oasis-open.org/legaldocml/ns/akn/3.0}authorialNote':
                            parent.remove(authorialNote)
        
                # Collect all text into a list
                text_values = [text for text in article.itertext() if text.strip()]
        
                # Join all child values into a single string
                content = ' '.join(text_values)
        
                # fill up the json data
                doc_data = {'heading': heading_content, 'original_name': original_name, 'name': name, 'content': content}
            
                # append the json data to the list
                data.append(doc_data)
        
        # convert the list to a json string
        json_data = json.dumps(data, indent=4, ensure_ascii=False)
        
        # Change the file extension from .xml to .json
        json_file = os.path.join(directory, os.path.splitext(filename)[0] + ".json")
        
        # write the json data to a file
        try: 
            with open(json_file, 'w', encoding='utf-8') as f:
                json.dump(data, f, indent=4, ensure_ascii=False)
        except Exception as e:
            print("An error occurred:", e)
        
    
```
